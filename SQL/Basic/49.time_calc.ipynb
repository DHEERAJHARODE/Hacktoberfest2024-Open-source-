```sql
select in_time, date_part('hour', in_time) 
    from guest 
    where in_time::time < '11:00'::time
```

## ช่วงเวลาที่ลูกค้าเข้ามารับประทานอาหาร

```sql
select date_part('hour', in_time)::int "period", count(*), 
    repeat('+', count(*)::int),
    repeat(chr(x'2588'::int), count(*)::int)
    from guest
    group by date_part('hour', in_time)
    order by date_part('hour', in_time)
```

## สมมติหากเป็นการคำนวณค่าบริการจอดรถ โดยคิดอัตรา 10 บาทต่อชั่วโมง โดยส่วนของนาที หากเกิน 20 นาทีจะคิดเป็น 1 ชั่วโมง

```sql
-- parking fee = 10/hr
select 
    order_id, in_time, out_time,
    out_time - in_time "parking duration", 
    date_part('hour', out_time - in_time) hours,
    date_part('minute', out_time - in_time) mins,
    date_part('hour', out_time - in_time) * 10 +  -- ชั่วโมงละ 10 บาท
        case
            when date_part('minute', out_time - in_time) > 20 then 10
            else 0                        -- ถ้าส่วนที่เป็นนาทีไม่เกิน 20 นาทีจะไม่คิดตัง
        end "parking fee"
    from guest
    limit 10;
```
