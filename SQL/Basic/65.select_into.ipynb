# select...into เพื่อนำไปสร้างตารางใหม่

```sql
select * from province
```

```sql
select p_id, name from province
```

```sql
select p_id, name
    into abbr
    from province;
```

```sql
select region, count(*) "# provinces", sum(area_km2) as area
	from province
	group by region;
```

```sql
select region, count(*) "# provinces", sum(area_km2) as area
	into province_infomation
	from province
	group by region;
```

```sql
select * 
	into buri
	from province
	where name like '%บุรี%'
```
