**Trignometric Graphs using numpys**

__1.SIN__

```python
x=np.arange(0,np.pi*3,0.5)
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.axhline(y=0,color='r',linestyle='-')

# plt.plot(x,y,"^")
plt.title("Sin Graph")
plt.xlabel("theta")
plt.ylabel("Value")
```

__2.TAN__

```python
x=np.linspace(-2*np.pi,2*np.pi,1000)
y=np.tan(x)
```

```python
plt.plot(x,y)
plt.ylim(-5,5)
```

__3.COSX__

```python
x=np.linspace(-2*np.pi,2*np.pi)
y=np.cos(x)
```

```python
plt.plot(x,y)

plt.axhline(y=0)
```

__4.COSECX__

```python
x=np.linspace(-2*np.pi,2*np.pi)
y=1/np.sin(x)
```

```python
plt.plot(x,y)
plt.ylim(-5,5)
plt.axhline(y=0)
```

__SECX__

```python
x=np.linspace(-2*np.pi,2*np.pi)
y=1/np.cos(x)
```

```python
plt.plot(x,y)
plt.ylim(-5,5)
plt.axhline(y=0)
```

__COTX__

```python
x=np.linspace(-2*np.pi,2*np.pi)
y=1/np.tan(x)
```

```python
plt.plot(x,y)
plt.axhline(y=0)
plt.ylim(-5,5)
```

__VERTICAL GRAPH__

```python
x=[2,5,7,9]
y=[3,2,4,5]

x1=[5,4,3,2]
y1=[1,7,5,3]
```

```python
plt.bar(x,y,color='g',linewidth=6)
plt.bar(x1,y1,color='r',linewidth=3)

plt.title("BAR graph")
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend(['lin1','line2'],loc=4)
```

__HORIZONTAL GRAPH__

```python
x=[2,5,7,9]
y=[3,2,4,5]

x1=[5,4,3,2]
y1=[1,7,5,3]
```

```python
plt.barh(x,y,color='g',linewidth=6)
plt.barh(x1,y1,color='r',linewidth=3)

plt.title("BAR graph")
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend(['line1','line2'],loc=4)
```
